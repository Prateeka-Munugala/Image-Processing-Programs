```python
#12. Implement erosion and dilation based morphology

import cv2
import numpy as np
import matplotlib.pyplot as plt

image=cv2.imread('/content/lena.jpg',0)

_,binary_img = cv2.threshold(image,127,255,cv2.THRESH_BINARY)
kernel=np.ones((3,3),np.uint8)

erosion = cv2.erode(binary_img,kernel,iterations=1)
dilation = cv2.dilate(binary_img,kernel,iterations=1)

titles = ['Original','Erosion','Dilation']
images = [binary_img,erosion,dilation]

for i in range(3):
  plt.subplot(1,3,i+1)
  plt.imshow(images[i],'gray')
  plt.title(titles[i])
  plt.xticks([]),plt.yticks([])
plt.show()
```
