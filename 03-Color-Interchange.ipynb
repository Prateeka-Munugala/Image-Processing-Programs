```python
#3. Python program using matplotlib to interchange RED and BLUE planes in each picture

import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread("/content/dice.jpg")

b,g,r=cv.split(img)
img_interchanged = cv.merge((r,g,b))

plt.figure(figsize=(6,6))

plt.subplot(121)
plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB))
plt.title("Original Image")

plt.subplot(122)
plt.imshow(cv.cvtColor(img_interchanged,cv.COLOR_BGR2RGB))
plt.title("Image after Red and Blue are interchanged")

plt.show()
```
