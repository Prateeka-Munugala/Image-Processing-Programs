```python
#14. Python program to implement face detection using feature matching techniques

import cv2
import numpy as np
import matplotlib.pyplot as plt

reference_img = cv2.imread("/content/1.jpg",0)
target_img = cv2.imread("/content/2.jpg",0)

orb = cv2.ORB_create()
keypoints_ref, descriptors_ref = orb.detectAndCompute(reference_img, None)
keypoints_target, descriptors_target = orb.detectAndCompute(target_img, None)

bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
matches = bf.match(descriptors_ref, descriptors_target)
matches = sorted(matches, key = lambda x:x.distance)

result_image = cv2.drawMatches(reference_img, keypoints_ref, target_img, keypoints_target,
                               matches[:50], None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

plt.figure(figsize=(10, 6))
plt.imshow(result_image)
plt.title('Face Detection using Feature Matching')
plt.show()
```
