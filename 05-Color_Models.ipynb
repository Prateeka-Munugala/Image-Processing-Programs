```python
#5. Python program in analysing the images using different color models

import cv2
import matplotlib.pyplot as plt

def load_image(image_path):
    img=cv2.imread(image_path)
    img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    return img_rgb
def convert_to_hsv(img_rgb):
    img_hsv=cv2.cvtColor(img_rgb,cv2.COLOR_RGB2HSV)
    return img_hsv
def convert_to_lab(img_rgb):
    img_lab=cv2.cvtColor(img_rgb,cv2.COLOR_RGB2LAB)
    return img_lab
def plot_image(img,title,color_map=None):
    plt.imshow(img,cmap=color_map)
    plt.title(title)
    plt.axis("off")
def analyze_image(image_path):
    img_rgb=load_image(image_path)
    img_hsv=convert_to_hsv(img_rgb)
    img_lab=convert_to_lab(img_rgb)
    plt.figure(figsize=(18,6))
    plt.subplot(1,3,1)
    plot_image(img_rgb,"RGB Color Model")
    plt.subplot(1,3,2)
    plot_image(img_hsv,"HSV Color Model")
    plt.subplot(1,3,3)
    plot_image(img_lab,"LAB Color Model")
    plt.show()
image_path="/content/flower.jpg"
analyze_image(image_path)
```
