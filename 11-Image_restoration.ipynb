```python
#11. Implementation of image restoring techniques
import cv2
import numpy as np
import matplotlib.pyplot as plt

noisy_image = cv2.imread('/content/flower.jpg',cv2.IMREAD_GRAYSCALE)
restored_image = cv2.fastNlMeansDenoising(noisy_image,None,h=10,templateWindowSize=7,searchWindowSize=21)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(noisy_image,cmap='gray')
plt.title('Noisy Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(restored_image,cmap='gray')
plt.title('Restored Image')
plt.axis('off')

plt.show()
```
