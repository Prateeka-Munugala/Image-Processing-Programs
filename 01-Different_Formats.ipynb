```python
#1. Program to acquire an image, store in different formats and display the properties of the images

from PIL import Image
import matplotlib.pyplot as plt
image_path = "/content/flower.jpg"
img=Image.open(image_path)
formats=["JPEG","PNG","BMP"]
for fmt in formats:
  img.save(f"output.{fmt.lower()}",fmt)

for fmt in formats:
    with Image.open(f"output.{fmt.lower()}") as img:
        print(f"Format: {img.format}")
        print(f"Size: {img.size}")
        print(f"Mode: {img.mode}")
        plt.imshow(img)
        plt.title(f"Image in {img.format} format")
        plt.axis("off")
        plt.show()
        print()
```
