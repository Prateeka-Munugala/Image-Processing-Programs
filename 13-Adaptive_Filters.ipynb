```python
#13. Apply adaptive filters to remove from the image

import cv2
import numpy as np
import matplotlib.pyplot as plt

def add_gaussian_noise(image,mean=0,var=5000):
  sigma=var**0.5
  gaussian=np.random.normal(mean,sigma,image.shape)
  noisy_image=np.clip(image+gaussian,0,255)
  return noisy_image.astype(np.uint8)

img = cv2.imread("/content/ball.jpg")
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
noisy_image=add_gaussian_noise(gray_img)
filtered_image =cv2.bilateralFilter(noisy_image,d=15,sigmaColor=150,sigmaSpace=150)

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(gray_img,cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(noisy_image,cmap='gray')
plt.title('Noisy Image')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(filtered_image,cmap='gray')
plt.title('Filtered Image(Adaptive Bilateral)')
plt.axis('off')

plt.show()
```
